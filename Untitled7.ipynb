```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(102316085)


sales_data = np.random.randint(1000, 5000, size=(12, 4))


categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame(sales_data, columns=categories, index=months)


df
```

```python

print("First 5 Rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())


category_totals = df.sum()
monthly_totals = df.sum(axis=1)

print("\nTotal Sales per Category:\n", category_totals)
print("\nTotal Sales per Month:\n", monthly_totals)


growth = df.pct_change().mean() * 100
print("\nAverage Sales Growth per Category (%):\n", growth)


df['Total Sales'] = df.sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100
df
```

```python

df['Electronics_Discounted'] = df['Electronics'] * 0.9
df
```

```python

plt.figure(figsize=(12, 6))
for col in categories:
    plt.plot(df.index, df[col], label=col)
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution per Category')
plt.ylabel('Units Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```


Q2: Array Calculations


```python

array = np.array([[1, -2, 3], [-4, 5, -6]])


abs_array = np.abs(array)


flat = array.flatten()
percentiles_flat = np.percentile(flat, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)


stats = {
    'Flattened': [flat.mean(), np.median(flat), flat.std()],
    'Columns': [(array[:,i].mean(), np.median(array[:,i]), array[:,i].std()) for i in range(array.shape[1])],
    'Rows': [(array[i,:].mean(), np.median(array[i,:]), array[i,:].std()) for i in range(array.shape[0])]
}

abs_array, percentiles_flat, percentiles_col, percentiles_row, stats
```

Q3: NumPy Rounding

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floors = np.floor(a)
ceilings = np.ceil(a)
truncated = np.trunc(a)
rounded = np.round(a)

floors, ceilings, truncated, rounded
```

Q4: Swap Two Elements in a List Using a Temporary Variable


```python
lst = [10, 20, 30, 40]
i, j = 1, 3

temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

lst
```

Q5: Swap Two Elements in a Set by Converting to a List


```python
s = {100, 200, 300, 400}
lst_s = list(s)

i, j = 0, 2
lst_s[i], lst_s[j] = lst_s[j], lst_s[i]

s_swapped = set(lst_s)
s_swapped
```
